```python
queries = [
'смотреть сериалы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт',
#'сериалы онлайн смотреть бесплатно'
]

a = []
for i in queries:
    a.append(len(i.split()))

c_a = {}
for i in a:
    c_a[i] = a.count(i)
c_a = list(c_a.items())

for i in range(0, len(c_a)):
    print('Поисковых запросов, содержащих', c_a[i][0], 'слов(а):', round((c_a[i][1]/len(queries))*100, 2), '%')
```

```python
results = {
'vk': {'revenue': 103, 'cost': 98},
'yandex': {'revenue': 179, 'cost': 153},
'facebook': {'revenue': 103, 'cost': 110},
'adwords': {'revenue': 35, 'cost': 34},
'twitter': {'revenue': 11, 'cost': 24},
}

a_key = list(results.keys())       ## Разбиваем на состовляющие
a_znah = list(results.values())
b_znah = []
b_key = []
for i in a_znah:
    b_znah.append(list(i.values()))
    b_key.append(list(i.keys()))
    
for i in b_znah:                   ## Считаем нужное значение и вносим к остальным
    x = round((((i[0] / i[1]) -1) * 100), 2)
    i.append(x)

for i in b_key:                    ## Добавляем ключ под подсчитаное значение
    i.append('ROI')

c = {}
a = {}
for i in range(len(a_key)):         
    for j in range(len(b_key[i])): 
        c[b_key[i][j]] = b_znah[i][j]
    a[a_key[i]] = c 
    
print(a)
```
