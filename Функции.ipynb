```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def poisc_documenta(num_doc):                 ## Вспомагательные функции
    for i in documents:
        if i['number'] == num_doc:
            return True

def poisc_polki(num_polki):
    for i in list(directories.keys()):
        if num_polki == i:
            return True
```

```python
def poisc_vladelca_dokumenta(num_doc):
    if poisc_documenta(num_doc) == True:
        for i in documents:
            if i['number'] == num_doc:
                flag = i['name']
    else:
        flag = 'Документ не найден.'
    return flag
```

```python
def poisc_nomera_polki(num_doc):
    flag = 'Документ не найден.'
    for i in range(1, len(directories) +1):
        c = directories[str(i)]
        for j in c:
            if j == num_doc:
                flag = str(i)
    return flag
```

```python
def informacia_po_vsem_dokumentam():
    for i in documents:
        print('№',i['number'],', тип:',i['type'],', имя владельца:',i['name'],', номер полки:',poisc_nomera_polki(i['number']))
```

```python
def dobavlenie_novoi_polki(num_polki):
    if poisc_polki(num_polki) == True:
        flag = 'Такая полка уже существует. Текущий перечень полок:'
    else:
        directories[num_polki] = []
        flag = 'Полка добавлена. Текущий перечень полок:'
    return flag
```

```python
def ydalenie_polki(num_polki):
    if poisc_polki(num_polki) != True:
        flag = 'Такой полки не существует. Текущий перечень полок:'
    if poisc_polki(num_polki) == True:
        if len(directories[num_polki]) == 0:
            del directories[num_polki]
            flag = 'Полка удалена. Текущий перечень полок:'
        else:
            flag = 'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:'
    return flag
```

```python
def dobavlenie_novogo_documenta(num_doc,tipe_doc,name_vladelca,num_polki):
    if poisc_documenta(num_doc) == True:
        flag = 'Документ уже существует.'
    if poisc_polki(num_polki) == True:
        directories[str(num_polki)].append(num_doc)
        flag = 1
    else:
        flag = 'Такой полки не существует. Добавьте полку командой ads.'
    if flag == 1:
        documents.append({'type': tipe_doc, 'number': num_doc, 'name': name_vladelca})
        flag = 'Документ добавлен.'
    return flag
```

```python
def ydalenie_documenta(num_doc):
    if poisc_documenta(num_doc) == True:
        for i in documents:
            if i['number'] == num_doc:
                documents.remove(i)
        for i in range(1, len(directories) +1):
            c = directories[str(i)]
            for j in c:
                if j == num_doc:
                    directories[str(i)].remove(num_doc)
                    flag = 'Документ удалён.'
    else:
        flag = 'Документ не найден.'
    return flag
```

```python
def peremechenie_dokumenta(num_doc,num_polki):
    if poisc_documenta(num_doc) == True:
        if poisc_polki(num_polki) == True:
            flag = 1
        else:
            flag = 'Такой полки не существует.' 
    else:
        flag = 'Документ не найден.'
    if flag == 1:
        for i in range(1, len(directories) +1):
            d = directories[str(i)]
            for j in d:
                if j == num_doc:
                    directories[str(i)].remove(num_doc)
        for i in list(directories.keys()):
            if num_polki == i:
                directories[str(num_polki)].append(num_doc)
                flag = 'Документ перемещён.'
    return flag
```

```python
def comand():
    print('р - узнать владельца документа, s - узнать номер полки, l - полная информация по всем документам,', 
          'ads - добавить новую полку, ds - удалить существующую полку, ad - добавить новый документ,',
          'd - удалить документ, m - переместить документ, q - закончить ввод.')
    comanda = 0
    razdelitel = ','
    while comanda != 'q':
        comanda = input('Введите команду:')
        match comanda:
            case 'p':
                num_doc = input('Введите номер документа')
                if poisc_vladelca_dokumenta(num_doc) != 'Документ не найден':
                    print('Имя владельца:', poisc_vladelca_dokumenta(num_doc))
                else:
                    print(poisc_vladelca_dokumenta(num_doc))
            case 's':
                num_doc = input('Введите номер документа')
                if poisc_nomera_polki(num_doc) != 'Документ не найден':
                    print('Номер полки:', poisc_nomera_polki(num_doc))
                else:
                    print(poisc_nomera_polki(num_doc))
            case 'l':
                informacia_po_vsem_dokumentam()
            case 'ads':
                num_polki = input('Введите номер полки:')
                print(dobavlenie_novoi_polki(num_polki), razdelitel.join(list(directories.keys())))
            case 'ds':
                num_polki = input('Введите номер полки:')
                print(ydalenie_polki(num_polki), razdelitel.join(list(directories.keys())))
            case 'ad':
                num_doc = input('Введите номер документа:')
                tipe_doc = input('Введите тип документа:')
                name_vladelca = input('Введите владельца документа:')
                num_polki = input('Введите номер полки:')
                print(dobavlenie_novogo_documenta(num_doc,tipe_doc,name_vladelca,num_polki), 'Текущий список документов:')
                informacia_po_vsem_dokumentam()
            case 'd':
                num_doc = input('Введите номер документа')
                print(ydalenie_documenta(num_doc), 'Текущий список документов:')
                informacia_po_vsem_dokumentam()
            case 'm':
                num_doc = input('Введите номер документа:')
                num_polki = input('Введите номер полки:')
                print(peremechenie_dokumenta(num_doc,num_polki), 'Текущий список документов:')
                informacia_po_vsem_dokumentam()
   
comand()
```
