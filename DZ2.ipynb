```python
a = str(input('Введите слово'))

len_a = len(a)
if len_a % 2 != 0:
    inds_a = len(a) // 2 
    print(a[inds_a])
else:
    inds_a1 = len(a) // 2 - 1
    inds_a2 = len(a) // 2
    print(a[inds_a1]+a[inds_a2])
```

```python
boys = []
girls = []

b = str(input('Введите имя парня, чтобы перейти к девушкам введите 0'))
while b != '0':
    boys.append(b)
    b = str(input('Введите имя парня, чтобы перейти к девушкам введите 0'))
    
g = str(input('Введите имя девушки, чтобы прекратить ввод введите 0'))
while g != '0':
    girls.append(g)
    g = str(input('Введите имя девушки, чтобы прекратить ввод введите 0'))

if len(boys) == len(girls):
    boys = sorted(boys)
    girls = sorted(girls)
    print('Идеальные пары:')
    while len(boys) != 0:
        print(boys[0], girls[0])
        del(boys[0])
        del(girls[0])
else:
    print('Внимание! Кто-то может остаться без пары!')
```

```python
countries_temperature = [
['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['Германия', [57.2, 55.4, 59, 59, 53.6]],
['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

for i in countries_temperature:
    temp_c = []
    while len(i[1]) != 0:
        temp_c.append((i[1][0] - 32) * 5 / 9)
        del(i[1][0])
    sr_temp = round(sum(temp_c) / len(temp_c),1)
    print(i[0], 'Средняя температура -', sr_temp)
```

```python
import re
car_ids = ['А222ВС96', 'АБ22ВВ193']
pattern = r'([АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})'

for i in car_ids:
    match = re.match(pattern, i)
    if match is None:
        print('Номер', i, 'не валиден')
    elif match is not None:
        print('Номер валиден', match.group(1),', регион -', match.group(2))
```

```python
stream = [
'user4,2021-01-01;3',
'user3,2022-01-07;4',
'user2,2022-03-29;1',
'user1,2020-04-04;13',
'user2,2022-01-05;7',
'user1,2021-06-14;4',
'user3,2022-07-02;10',
'user4,2021-03-21;19',
'user4,2022-03-22;4',
'user4,2022-04-22;8',
'user4,2021-05-03;9',
'user4,2022-05-11;11'
]

stream1 = [
'user100,2022-01-01;150',
'user99,2022-01-07;205',
'user1001,2022-03-29;81'
]

import re

us = []
v = []

for i in stream1:
    a = re.split(r";|,", i)
    us.append(a[0])
    v.append(int(a[2]))
    
sum_v = sum(v)
count_us = len(list(set(us)))
sr_v = round(sum_v / count_us, 2)
print('Среднее колличество просмотров на пользователя:', sr_v)
```
