```python
pover = pd.read_csv('C:/Users/sanar/Desktop/10._Основы_pandas/10._Основы_pandas/Файлы к заданию №2/power.csv')
pover.head()
```

```python
pover = pover[pover['country'].isin(['Latvia', 'Lithuania', 'Estonia'])]
pover.head()
```

```python
pover = pover[pover['category'].isin([4, 12, 21])]
pover.head()
```

```python
pover = pover[(pover['year'] >= 2005) & (pover['year'] <= 2010)]
pover.head()
```

```python
pover = pover[pover['quantity'] > 0]
pover.head()
```

```python
pover = pover.groupby('country').sum('quantity')
pover.head()
```
