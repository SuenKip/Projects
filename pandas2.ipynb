```python
join = rating.merge(move, on='movieId', how='left')
join.head()
```

```python
def ocenka(rat):
    if rat >= 4.5:
        return 'высокий рейтинг'
    elif rat > 2 and rat <= 4:
        return 'средний рейтинг'
    else:
        return 'низкий рейтинг'
```

```python
join['class'] = join['rating'].apply(ocenka)
join.head()
```

```python
keywords = pd.read_csv('C:/Users/sanar/Desktop/ml-latest-small/ml-latest-small/keywords.csv')
keywords.head()
```

```python
def geo(key):
    if 'москва' in key:
        return 'Центр'
    elif 'тула' in key:
        return 'Центр'
    elif 'ярославль' in key:
        return 'Центр'
    elif 'петербург' in key:
        return 'Северо-Запад'
    elif 'псков' in key:
        return 'Северо-Запад'
    elif 'мурманск' in key:
        return 'Северо-Запад'
    elif 'владивосток' in key:
        return 'Дальний Восток'
    elif 'сахалин' in key:
        return 'Дальний Восток'
    elif 'хабаровск' in key:
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo)
keywords.head()
```

```python
keywords = keywords[keywords['region'].isin(['Северо-Запад', 'Дальний Восток', 'Центр'])]
keywords.head()
```
