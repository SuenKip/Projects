```python
log = pd.read_csv('C:/Users/sanar/Desktop/доп_материалы_к_пандас_1_и_2/visit_log.csv', sep = ';')
log.head()
```

```python
def type(per):
    traffic_source = per['traffic_source']
    region = ['region']
    if traffic_source == 'yandex' or traffic_source == 'google':
        return 'organic'
    elif (region == 'Russia') and (traffic_source == 'paid' or traffic_source == 'email'):
        return 'ad'
    elif (region != 'Russia') and (traffic_source == 'paid' or traffic_source == 'email'):
        return 'other'
    else:
        return traffic_source
```

```python
log['source_type'] = log.apply(type, axis=1)
log.head()
```

```python
ratings = pd.read_csv('C:/Users/sanar/Desktop/доп_материалы_к_пандас_1_и_2/ml-latest-small/ratings.csv')
ratings.head()
```

```python
users = ratings.groupby('userId').agg({'userId':'count','timestamp':['min','max']})
users.head()
```

```python
users = users[users[('userId', 'count')] > 100]
users.head()
```

```python
users['vremia_gzizni'] = users.timestamp['max'] - users.timestamp['min']
users.head()
```

```python
users['vremia_gzizni'].mean()
```
