```python
a = str(input('Введите первую стороку'))
b = str(input('Введите вторую строку'))

len_a = len(a)
len_b = len(b)

if len_a > len_b:
    print('Первая строка длинее второй')
elif len_b > len_a:
    print('Вторая строка длинее первой')
else:
    print('Строки равны')
```

```python
g = int(input('Введите год'))

if g % 4 == 0:
    if g % 100 == 0:
        if g % 400 == 0:
            print('Год високосный')
        else:
            print('Год не високосный')
    else:
        print('Год високосный')
else:
    print('Год не високосный')
```

```python
day = int(input('Введите день своего рождения'))
mount = str(input('Введите название месяца своего рождения'))

if (mount == 'Март' and day >= 21) or (mount == 'Апрель' and day <= 20):
    print('Овен') 
elif (mount == 'Апрель' and day >= 21) or (mount == 'Май' and day <= 20):
    print('Телец')
elif (mount == 'Май' and day >= 21) or (mount == 'Июнь' and day <= 21):
    print('Близнецы')
elif (mount == 'Июнь' and day >= 22) or (mount == 'Июль' and day <= 22):
    print('Рак')
elif (mount == 'Июль' and day >= 23) or (mount == 'Август' and day <= 22):
    print('Лев')
elif (mount == 'Август' and day >= 23) or (mount == 'Сентябрь' and day <= 23):
    print('Дева')
elif (mount == 'Сентябрь' and day >= 24) or (mount == 'Октябрь' and day <= 23):
    print('Весы')
elif (mount == 'Октябрь' and day >= 24) or (mount == 'Ноябрь' and day <= 22):
    print('Скорпион')
elif (mount == 'Ноябрь' and day >= 23) or (mount == 'Декабрь' and day <= 21):
    print('Стрелец')
elif (mount == 'Декабрь' and day >= 22) or (mount == 'Январь' and day <= 20):
    print('Козерог')
elif (mount == 'Январь' and day >= 21) or (mount == 'Февраль' and day <= 18):
    print('Водолей')
else:
    print('Рыбы')
```

```python
d = int(input('Введите длинну'))
v = int(input('Введите высоту'))
h = int(input('Введите ширину'))

if d <= 15 and v <= 15 and h <= 15:
    print('Коробка №1')
elif d > 200 or v > 200 or h > 200:
    print('Упаковка для лыжь')
elif (d > 15 and d < 50) or (v > 15 and v < 50) or (h > 15 and h < 50):
    print('Коробка №2')
else:
    print('Коробка №3')
```

```python
b = int(input('Введите номер билета'))
b1 = 0
b2 = 0
c = 6

while c > 3:
    b2 += b % 10
    b = b // 10
    c -= 1
while c > 0:
    b1 += b % 10
    b = b // 10
    c -= 1
    
if b1 == b2:
    print('Счастливый билет')
else:
    print('Несчастливый билет')
```

```python
a = str(input('Введите Круг, Треугольник или Прямоугольник'))
import math

if a == 'Круг':
    r = int(input('Введите радиус'))
elif a == 'Треугольник':
    t1 = int(input('Введите первую сторону'))
    t2 = int(input('Введите вторую сторону'))
    t3 = int(input('Введите третью сторону'))
else:
    p1 = int(input('Введите первую сторону'))
    p2 = int(input('Введите вторую сторону'))

if a == 'Круг':
    s = math.pi * (r ** 2)
    print(s)
elif a == 'Треугольник':
    p = (t1 + t2 + t3) / 2
    s = math.sqrt(p*(p-t1)*(p-t2)*(p-t3))
    print(s)
else:
    s = p1 * p2
    print(s)
```
